```python
import json

import numpy as np
from main_trainer import SQL2Circuits

# Reasonable results so far
test_case_folders = [
    "1_pennylane_optax_state_cardinality_25_25_007_4",
    "5_pennylane_optax_state_cost_25_25_007_2",
    "8_pennylane_optax_state_Sim14Ansatz_cardinality_10_20_007_8",
    "9_pennylane_optax_adamw_state_Sim14Ansatz_execution_time_50_50_007_4"]

for fold in test_case_folders:
    print(fold)
    training_config = None
    with open(f"./training/results/{fold}/training_stats.json", "r") as f:
        training_config = json.load(f)
    run_id = training_config["run_id"]
    if "circuit_architecture" in training_config:
        circuit_architecture = training_config["circuit_architecture"]
    else:
        circuit_architecture = "IQPAnsatz"
    classification =  training_config["classification"]
    classification = int(np.log2(classification))
    seed_file = training_config["seed_file"]
    qc_framework = training_config["qc_framework"]
    classical_optimizer = training_config["classical_optimizer"]
    measurement = "iqm"
    workload_type = training_config["workload_type"]

    model = SQL2Circuits(run_id = run_id,
                        classification = classification,
                        circuit_architecture = circuit_architecture,
                        seed_file = seed_file, 
                        qc_framework = qc_framework, 
                        classical_optimizer = classical_optimizer, 
                        measurement = measurement, 
                        workload_type = workload_type, 
                        initial_number_of_circuits = None, 
                        number_of_circuits_to_add = None,
                        iterative = None,
                        epochs = None,
                        learning_rate=None,
                        identifier=fold)

    model.evaluate_on_IQM()
```
